### Data lista de Tecnolite para ser guardada en Firestore

```python
import json
import firebase_admin
from firebase_admin import credentials
from firebase_admin import firestore

# Use a service account Voltz DEV
#cred = credentials.Certificate('secrets/clave-servicio-voltz-dev.json')

# Use a service account Voltz Pro
cred = credentials.Certificate('secrets/voltz-pro-key.json')

app = firebase_admin.initialize_app(cred)

db = firestore.client()
```

```python

# Opening JSON file
f = open('tecnolite_to_firestore_v4.json')
# returns JSON object as 
# a dictionary
data = json.load(f)
```

```python
batch = db.batch()
i = 0
t = 0
print('Total registros a guardar: ', len(data))

for x in range(len(data)):
    ref = db.collection(u'products').document(data[x]['catalog_code'] + '_' + data[x]['sku'].replace('/', '_').replace(' ', '_'))
    data_with_timestamp = data[x].copy()
    data_with_timestamp['updated_at'] = firestore.SERVER_TIMESTAMP  # Add the server timestamp
    batch.set(ref, data_with_timestamp, merge=True)  # Use set with merge option
    t += 1
    
    if i == 10 or len(data) == t:
        batch.commit()
        i = 0
        batch = db.batch()
        print('Guardados: ' + str(t) + ' de ' + str(len(data)))
    else: 
        i += 1

print('Total de registros guardados', t)
```
